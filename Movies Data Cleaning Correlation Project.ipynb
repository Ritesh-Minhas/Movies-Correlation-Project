```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the Dataset

df = pd.read_csv(r"C:\Users\cd\Desktop\Pandas\movies.csv")
df
```

```python
# Exploring the Dataset
df.info()
```

```python
# Removing Duplicates (if any)
df = df.drop_duplicates()
df
```

```python
# Checking the number of null values
df.isnull().sum()
```

```python
# Correcting the year column, which matches the released column's year
df['released'].str.split(",", 1, expand = True)
df[["Date", "CorrectYear"]] = df['released'].str.split(",", 1, expand = True)
df
```

```python
# Corrected the Year
x = df['CorrectYear'].str.split("(", 1, expand = True)
df['CorrectedYear'] = x[0]
df
```

```python
# Dropping the Null Values
df = df.dropna()
df
```

```python
# Dropping the useless column
df = df.drop(columns = ["Date","CorrectYear"])
df
```

```python
# Resetting the index
df.reset_index(drop = True, inplace = True)
df
```

```python
# Converting the budget and gross column data type to int64
df['budget'] = df['budget'].astype("int64")
df['gross'] = df['gross'].astype("int64")
df.info()
```

```python
# Sorting the dataset by gross column
df = df.sort_values(by = 'gross', ascending = False)
df
```

```python
# Using scatter plot to see the correlation between budget and gross column

plt.figure(figsize = (12,5))
plt.scatter(df['budget'], df['gross'], c = "green")
plt.title("Budget vs Gross")
plt.xlabel("Gross")
plt.ylabel("Budget")
plt.show()
```

```python
# Plotting Budget vs Gross using seaborn
plt.figure(figsize = (12,5))
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color" : "green"}, line_kws = {"color" : "blue"})

plt.title("Budget vs Gross")
plt.xlabel("Gross")
plt.ylabel("Budget")
plt.show()
```

```python
# Finding the correlation

# Converting the CorrectedYear data type to int64
df['CorrectedYear'] = df['CorrectedYear'].astype('int64')

df.corr(method = 'pearson') # method = 'pearson', 'kendall', 'spearman'
```

```python
corr_matrix = df.corr(method = 'pearson')

# PLotting the correlation matrix

plt.figure(figsize = (12,8))
sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation Matrix")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# Converting the object data types to numeric values for plotting

df_numerized = df

for col in df_numerized.columns:
    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes

df_numerized
```

```python
corr_matrix = df_numerized.corr(method = 'pearson')

# PLotting the correlation matrix

plt.figure(figsize = (12,8))
sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation Matrix")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# For finding correlation between every field with other (in form of Series)

corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack() # Converts the DataFrame into Series
corr_pairs
```

```python
# Sorting the correlation

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Selecting correlations with values greater than 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
